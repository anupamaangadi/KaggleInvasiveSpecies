# Now try and create our first model that will sit on the end of the conv layers
## A couple of fully connected layers with Batch Norm and Dropout

```python
def get_bn_layers(p):
    return [
        MaxPooling2D(input_shape=conv_layers[-1].output_shape[1:]),
        BatchNormalization(axis=1),
        Dropout(p/4),
        Flatten(),
        Dense(512, activation='relu'),
        BatchNormalization(),
        Dropout(p),
        Dense(512, activation='relu'),
        BatchNormalization(),
        Dropout(p/2),
        Dense(2, activation='softmax')
    ]
```

```python
p=0.6
```

```python
bn_model = Sequential(get_bn_layers(p))
bn_model.compile(Adam(lr=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
bn_model.fit(conv_feat, trn_labels, batch_size=batch_size, nb_epoch=3, 
             validation_data=(conv_val_feat, val_labels))
```

```python
bn_model.save_weights(path+'models/conv_512_6_v2.h5')
```

```python
bn_model.optimizer.lr = 1e-4
bn_model.fit(conv_feat, trn_labels, batch_size=batch_size, nb_epoch=1, 
             validation_data=(conv_val_feat, val_labels))
```

```python
bn_model.load_weights(path+'models/conv_512_6_v2.h5')
```

```python
bn_model.evaluate(conv_val_feat, val_labels)
```
